```python
loan_df = loan_df.drop(['CUST_ID','random','NO_OF_L_DR_TXNS','NO_OF_L_CR_TXNS','ACC_OP_DATE','AGE_BKT'],axis=1)
```

```python
loan_df.head(25)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split
```

```python
dummy_var = pd.get_dummies(loan_df['GENDER'],drop_first=True)
dummy_var1 = pd.get_dummies(loan_df['OCCUPATION'],drop_first=True)
dummy_var2 = pd.get_dummies(loan_df['ACC_TYPE'], drop_first=True)
loan_df_new = pd.concat([loan_df,dummy_var,dummy_var1,dummy_var2], axis=1)
loan_df_new = loan_df_new.drop(['GENDER','OCCUPATION','ACC_TYPE'], axis=1)
loan_df_new.columns
```

```python
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1)
x=loan_df_new[['AGE', 'BALANCE', 'SCR', 'HOLDING_PERIOD',
       'LEN_OF_RLTN_IN_MNTH', 'TOT_NO_OF_L_TXNS', 'NO_OF_BR_CSH_WDL_DR_TXNS',
       'NO_OF_ATM_DR_TXNS', 'NO_OF_NET_DR_TXNS', 'NO_OF_MOB_DR_TXNS',
       'NO_OF_CHQ_DR_TXNS', 'FLG_HAS_CC', 'AMT_ATM_DR', 'AMT_BR_CSH_WDL_DR',
       'AMT_CHQ_DR', 'AMT_NET_DR', 'AMT_MOB_DR', 'AMT_L_DR',
       'FLG_HAS_ANY_CHGS', 'AMT_OTH_BK_ATM_USG_CHGS', 'AMT_MIN_BAL_NMC_CHGS',
       'NO_OF_IW_CHQ_BNC_TXNS', 'NO_OF_OW_CHQ_BNC_TXNS', 'AVG_AMT_PER_ATM_TXN',
       'AVG_AMT_PER_CSH_WDL_TXN', 'AVG_AMT_PER_CHQ_TXN', 'AVG_AMT_PER_NET_TXN',
       'AVG_AMT_PER_MOB_TXN', 'FLG_HAS_NOMINEE', 'FLG_HAS_OLD_LOAN', 'M', 'O',
       'SAL', 'SELF-EMP', 'SENP', 'SA']]
y=loan_df_new['TARGET']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
clf.fit(x_train, y_train)
```

```python
pred = clf.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```
